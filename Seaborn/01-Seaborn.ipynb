# Seaborn
## Why Seaborn?
* provides a layer of abstraction hence simpler to use
* better aesthetics
* more graphs included
## Why not just use seaborn?
| Limitation in Seaborn                           | Why Matplotlib Is Needed                                                             |
| ----------------------------------------------- | ------------------------------------------------------------------------------------ |
| Seaborn covers only **common plot types**       | Matplotlib supports **everything** from simple to complex, custom plots              |
| Less control over fine-tuning                   | Matplotlib gives **full control**: tick labels, legends, subplots, annotations, etc. |
| Seaborn doesn’t support **interactive plots**   | Matplotlib integrates with tools like **mplcursors**, **widgets**, **3D**, etc.      |
| Seaborn has limited **animation or 3D support** | Matplotlib supports **3D plotting** (`mpl_toolkits`) and **animations**              |
| Seaborn can’t plot from raw NumPy arrays easily | Matplotlib works directly with NumPy, which is crucial for some ML workflows         |

### Facet Plot
A facet plot (also called faceted plot or small multiples) is a way to split your data into subsets and plot each subset in its own small plot (facet), arranged in rows or columns.

**In Simple Terms:**

*Facet plot means "many small plots"—each showing a part of the data, like one plot per country, one per gender, or one per category.*

## 2. Distribution Plots

- Used for **univariate analysis**
- Helps understand the **distribution** of data
- Shows the **range** of observations
- Reveals **central tendency**  
  → Where is the center of the data?

- Checks for **modality**:
  - Is the data **multimodal**?
    - A **mode** is the value (or range) that appears most often.
    - A **multimodal distribution** has **two or more distinct peaks** (clusters).
    - Examples:
      - **Unimodal** → 1 peak
      - **Bimodal** → 2 peaks
      - **Trimodal** → 3 peaks

- Identifies **outliers**:
  - Outliers are data points that are **significantly different** from the rest.
  - They lie far away from the main cluster and can **affect results and visualizations**.

### 📊 Plots under Distribution Plot

- `histplot`
- `kdeplot`
- `rugplot`

## b. kdeplot - Kernel Density Estimate
* A KDE plot is a smooth, continuous curve that estimates the probability density of a numeric variable — basically, it shows where the data is most concentrated.
* Rather than using discrete bins, a KDE plot smooths the observations with a Gaussian kernel, producing a continuous density estimate.
* Instead of showing the raw counts like a histogram, a KDE plot shows a smoothed version of the distribution.
* Think of it as a smoothed histogram with a curve.
* It helps you see the shape of the distribution — such as:
    * Peaks (modes)
    * Spread (variance)
    * Skewness
    * Gaps or clusters

```python
sns.kdeplot(data=tips,x="total_bill")
```

**Density in a KDE plot tells you how likely it is that a value from the dataset falls in a certain range — not an exact value, but a range around it.**
* KDE shows probability density, not probability.
* The value on the y-axis is not a probability, but a density.
* To get the actual probability of a range (e.g., between 40 and 60), you'd need to calculate the area under the curve in that interval.

**In most scenarios, KDE plot is preferred over histplot because it helps in predicting.**

```python
sns.displot(data=tips,x="total_bill",kind="kde")
```

* #### hue parameter

```python
sns.displot(data=tips,x="total_bill",kind="kde",hue="sex")
```

* #### fill parameter 
**default --> False**

```python
sns.displot(data=tips,x="total_bill",kind="kde",hue="sex",fill=True)
```

## c. Rugplot
* Plot marginal distributions by drawing ticks along the x and y axes.
* This function is intended to complement other plots by showing the location of individual observations in an unobtrusive way.
* It visualizes the distribution of data points without binning or smoothing.
* Often used to complement a histogram or KDE plot.
* Helps you see the raw data directly on the axis.

```python
sns.kdeplot(data=tips,x="total_bill")
sns.rugplot(data=tips,x="total_bill")
```

**More ticks (closer together) in an area → more data points → higher density in that range.**

### Bivariate Histogram
* A bivariate histogram is an extension of a regular histogram that shows the distribution of two variables at once — like a 2D histogram.
* Instead of bars, it uses a grid of rectangles (bins) where:
    * x-axis: values of the first variable
    * y-axis: values of the second variable
    * Color or height of each rectangle = how many data points fall into that bin

```python
sns.histplot(data=tips,x="total_bill",y="tip")
```

```python
sns.displot(data=tips,x="total_bill",y="tip",kind="hist",cmap="viridis",cbar=True)
```

### Bivariate KDE Plot?
* A bivariate KDE (Kernel Density Estimate) plot shows the smoothed joint distribution of two continuous variables — like a 2D version of the KDE curve.
* **Purpose:** It estimates where data points are densely clustered in a 2D space using smooth contour lines or heatmaps instead of raw counts like in a 2D histogram.
* a bivariate KDE plot smoothes the (x, y) observations with a 2D Gaussian.

```python
sns.kdeplot(x="total_bill", y="tip", data=tips)
```

```python
sns.displot(kind="kde",x="total_bill", y="tip", data=tips,fill=True,cmap="inferno",cbar=True)
```

**seaborn.clustermap() uses hierarchical agglomerative clustering behind the scenes.**

### Changing sizes in figure level function
**We can't change it using matplotlib like we used to do in axes level ones.**

#### height:
* Controls the height (in inches) of each facet (subplot).
* Does not affect the overall figure size directly unless there are multiple rows.

📌 Default: 5

**Example:** height=4 means each subplot will be 4 inches tall.

#### aspect:
* Controls the aspect ratio: width = aspect × height
* So, it adjusts how wide each plot is relative to its height.

📌 Default: 1 (means width = height)

**Example:** aspect=1.5 with height=4 → width becomes 6 inches.


```python
sns.displot(data=tips,x="total_bill",kind="kde",hue="sex",fill=True,height=7,aspect=2)
```
