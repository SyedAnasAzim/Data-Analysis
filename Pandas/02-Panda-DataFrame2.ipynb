### apply(Series and DataFrame)

.apply() is a very useful method that allows you to apply a function to each element, row, or column of a DataFrame or Series.

```python
points_df = pd.DataFrame(
    {
        '1st point':[(3,4),(-6,5),(0,0),(-10,1),(4,5)],
        '2nd point':[(-3,4),(0,0),(2,2),(10,10),(1,1)]
    }
)

points_df
```

```python
def euclidean(point):
    p1=point["1st point"]
    p2=point["2nd point"]
    return ((p1[0]-p2[0])**2+(p1[1]-p2[1])**2)**0.5
```

default is column(axis=0) but we sending rows(axis=1)

```python
points_df["Distance"] = points_df.apply(euclidean,axis=1)
```
