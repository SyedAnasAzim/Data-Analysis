# Panda Strings - Vectorized string operations

## What are vectorized operations

* Vectorized operations are operations that are applied to entire arrays or sequences (like lists, NumPy arrays, or Pandas Series) at once, without using explicit loops like for or while.

* They are faster and more concise, and they take advantage of underlying optimized C/Fortran code in libraries like NumPy and Pandas.

```python
#for example
a = np.array([1,2,3,4])
a * 4
```

```python
s = [1,2,3]
[j**i for i in s for j in s]
```

### Problem in vectorized operations in vanilla python

* In python, if we have none in our list then it can not handle it and give error.
* Another problem is list comprehension which is relatively slow.

```python
st = ["Iron Man","Spider Man","Superman","Batman",None]
[i.startswith("S") for i in st]
```

### How pandas solve this problem?

To cope with these problems pandas introduced their own vectroized operations which not just only take care of none problem but also are very fast on larger data sets and the reson behind this speed is that pandas is built on numpy and numpy is nuilt on C.

* Pandas provides a .str accessor for vectorized string operations on Series of strings without writing loops.
* Numpy provides a np.char for vectorized string operations on array of strings without writing loops.(but it doesnt slove the none prolem)

If there is None in a pandas series then it will ignore it,


```python
s = pd.Series(["Iron Man","Spider Man","Superman","Batman",None])
#.str -> string accessor
s.str.startswith("S")
#Fast and optimized
```

### Common Functions

#### 1.upper()/lower()/capitalize()/title()

```python
df["Name"].str.upper()
```

```python
df["Name"].str.lower()
```

```python
df["Name"].str.capitalize()
```

- **title function makes upper case first letter of every word**

```python
df["Name"].str.title()
```

#### 2.len()

Gives the lenght of every elemnet in the series in form of series hence vectorized

```python
df["Name"].str.len()
```

**Find the passenger with the biggest name**

```python
df["Name"][df["Name"].str.len() == df["Name"].str.len().max()]
```

#### 3.strip()

```python
df["Name"][df["Name"].str[-1] == " "]
```

```python
df["Name"][df["Name"].str[-1] == " "].values
```

```python
df["Name"].str.strip().loc[[15,857]].values
```

As, we can see that no more extra space in the strings

#### 4.split() -> get()

**Making a column for surname and title (Mr./Mrs)**

```python
df["Last Name"] = df["Name"].str.split(',').str[0] #extracting surnames
df
```

**n attribute** in split tells how many splits to do

```python
df["Name"].str.split(',').str[1].str.split(n=1)
```

**expand attribute** (boolean: default = False) -> If true, coverts the list into dataframe

```python
df["Name"].str.split(',').str[1].str.split(n=1,expand=True)
```

```python
df[["Title","First Name"]] = df["Name"].str.split(',').str[1].str.split(n=1,expand=True)
df
```

```python
df["Title"].value_counts()
```

#### 5.replace()

```python
df["Title"] = df["Title"].str.replace("Ms","Miss").str.replace("Mlle","Miss")
df["Title"].value_counts()
```

### Flitering

#### 1.startswith/endswith

```python
df[df["First Name"].str.startswith("A")]
```

```python
df[df["First Name"].str.endswith("A")]
```

#### 2.isdigit/isalpha

```python
df[df["First Name"].str.isdigit()]
```

### Advanced level filtering

#### 1.contains 

**Find whose first name is john (no case-sensitivity)**

we use a attribute case=False(which is default True) for no case-sensitivity 

```python
df[df["First Name"].str.contains("joHn",case=False)]
```

#### What is Regex?
* Regex (short for Regular Expression) is a powerful tool used to search, match, and manipulate text patterns.
* It's like a mini-language for finding and replacing text based on patterns, not just fixed strings.



**Find the peoples last names which starts and ends with vowel**

```python
df[df["Last Name"].str.contains("^[aeiouAEIOU].+[aeiouAEIOU]$")]
```
