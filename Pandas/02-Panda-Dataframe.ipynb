## Creating DataFrame

A DataFrame is a two-dimensional, tabular data structure in pandas (Python's data analysis library), similar to an Excel spreadsheet or an SQL table. It consists of rows and columns, where:

Each column can have a different data type (integer, float, string, etc.).

Each row has an associated index, which uniquely identifies it.

## Filtering the DataFrame

```python
ipl.head(2)
```

1.  Find all the final winners.

```python
ipl.loc[ipl["MatchNumber"] == "Final","WinningTeam"]
```

2. How many super overs finishes have occured?

```python
ipl.loc[ipl["SuperOver"] == "Y"].shape[0]
```

3. How many matches CSK won in Kolkata?

```python
ipl.loc[(ipl["WinningTeam"] == "Chennai Super Kings") & (ipl["City"] == "Kolkata")].shape[0]
```

In pandas, the bitwise AND operator (&) is used instead of the logical AND operator (and) because of the way pandas handles Boolean indexing.

and cannot be used for vectorized operations on pandas Series.

It works for single Boolean values, like True and False, but not for entire columns.

4. Toss winner is match winner in percentage

```python
round((ipl.loc[ipl["TossWinner"] == ipl["WinningTeam"]]).shape[0]*100/ipl.shape[0],2)
```

```python
movies.head(2)
```

5. Movies rating higher than 8 and voting > 10000

```python
movies.loc[(movies["imdb_rating"] > 8) & (movies["imdb_votes"] > 10000)].shape[0]
```

6. Action movies with rating higher than 7.5

```python
movies.loc[(movies["genres"].str.split('|').apply(lambda x: "Action" in x)) & (movies["imdb_rating"] > 7.5)]
```

## Important DataFrame Methods

### astype

changes the datatype of a column

```python
ipl.info()
```

```python
ipl["ID"] = ipl["ID"].astype("int32")
ipl.info()
```

here we saves 4 kb memory usage here,so, astype method could be used as optimization

In pandas, category is a data type used for efficient storage and faster operations on categorical data. It is particularly useful when a column has a limited set of unique values (categories), such as gender, city names, or product categories.

```python
ipl["Season"] = ipl["Season"].astype("category")
ipl.info()
```

```python
ipl[["Team1","Team2"]] = ipl[["Team1","Team2"]].astype("category")
ipl.info()
```
