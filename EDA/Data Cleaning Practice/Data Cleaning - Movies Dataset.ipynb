### 2. Messy Data

```python
import re
pattern = re.compile(
    r'^(?:\((?P<version>[IVXLCDM]+)\)\s*)?'   # version like (I), (II), etc.
    r'\(?(?P<start>\d{4})?'                   # start year
    r'(?:[–-](?P<end>\d{4}| ))?\)?'           # end year or blank
    r'(?:\s*(?P<type>TV Movie|TV Special|Video Game|Video|Short))?' # type
    , re.IGNORECASE
)

def parse_entry(entry):
    if pd.isna(entry) or entry.strip().lower() == "nan":
        return pd.Series([None, None, None, None])
    
    m = pattern.search(entry)
    if m:
        version = m.group("version")
        start = m.group("start")
        end = m.group("end")
        ent_type = m.group("type")

        # Handle blank end (ongoing series)
        if end == " ":
            end = "Present"

        # Apply logic for default type
        if not ent_type:
            if start and end:  # has a range
                ent_type = "Series"
            elif start and not end:  # just one year
                ent_type = "Movie"

        # For movies → endyear must stay None
        if ent_type == "Movie":
            end = None

        return pd.Series([version, start, end, ent_type])
    
    return pd.Series([None, None, None, None])
```

```python
new = movies["YEAR"].apply(parse_entry)
new.columns = ["Version","Start_Year","End_Year","Type"]
new
```

```python
def extract_director_and_stars(stars: str) -> pd.Series:
    if stars.startswith("Director"):
        try:
            director, actors = stars.split("|", 1)  # only split once
            return pd.Series([director.strip()[9:].strip(":"), actors.strip()[5:].strip(":")])  
        except ValueError:
             # If "|" is missing but Director exists
             return pd.Series([stars[9:].strip(":"), "Unknown"])
    elif stars.startswith("Star"):
        return pd.Series(["Unknown", stars.strip()[5:].strip(":")])
    
    elif stars.startswith("Unknown"):
        return pd.Series(["Unknown", "Unknown"])
    
Directors_and_Actors = movies["STARS"].apply(extract_director_and_stars)
```

```python
Directors_and_Actors.columns = ["Director/Directors","Actor/Actors"]
Directors_and_Actors
```

```python
movies = pd.concat([movies,Directors_and_Actors],axis=1).drop(columns="STARS")
```
