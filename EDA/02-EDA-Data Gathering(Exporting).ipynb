# Exporting 
* to csv
* to excel
* to html
* to json
* to sql

## 1. to CSV

```python
df = pd.read_csv("01/deliveries.csv")
df
```

```python
temp = df.groupby("batsman")["batsman_runs"].sum().reset_index()
temp
```

```python
temp.to_csv("Export/Batsman.csv")
```

***It loads data with index, if we doesn't want index then we can use index=False parameter***

```python
temp.to_csv("Export/Batsman with no index.csv",index=False)
```

```python
temp = pd.pivot_table(df,index="batsman",columns="bowling_team",values="batsman_runs",aggfunc="sum",fill_value=0)
temp["Total"] = temp.sum(axis=1)
temp
```

```python
temp.sort_values(by="Total",ascending=False,inplace=True)
temp
```

```python
temp.to_csv("Export/Batsman-against-each-team.csv")
```

## 2. To Excel 

```python
temp = df.groupby("batsman")["batsman_runs"].sum().reset_index()
temp
```

```python
temp.to_excel("Export/output.xlsx")
```

### sheet_name parameter

```python
temp.to_excel("Export/output.xlsx",sheet_name="batsman runs")
```

```python
temp1= pd.pivot_table(df,index="batsman",columns="bowling_team",values="batsman_runs",aggfunc="sum",fill_value=0)
temp1
```

### Multiple Sheets

```python
with pd.ExcelWriter("Export/output.xlsx") as writer:
    temp.to_excel(writer,sheet_name = "batsman runs".title())
    temp1.to_excel(writer,sheet_name = "batsman runs against each team".title())
```
