# Matplotlib

* Matplotlib is a powerful data visualization library in Python. It is used to create a wide range of static, animated, and interactive plots, charts, and figures.
* Many popular Python visualization libraries are either built on top of Matplotlib or inspired by its architecture.

## Things to keep in mind while plotting graphs:
* Type of the Data
* How many variable are there

## Types of Data
* Categorical Data
* Numerical Data

## 1. Qualitative Data (Categorical Data):
Describes qualities or categories, not numbers.
| Type        | Description                 | Example                                                                           |
| ----------- | --------------------------- | --------------------------------------------------------------------------------- |
| **Nominal** | No natural order            | Gender (Male, Female), Colors (Red, Blue), Cities                                 |
| **Ordinal** | Has a natural order or rank | Education level (High school < Bachelor's < Master's), Ratings (Poor, Fair, Good) |


## 2. Quantitative Data (Numerical Data):
Describes quantities or measurable values.
| Type           | Description                                      | Example                                   |
| -------------- | ------------------------------------------------ | ----------------------------------------- |
| **Discrete**   | Countable values                                 | Number of students, Cars in a parking lot |
| **Continuous** | Measurable and can take any value within a range | Height, Weight, Temperature, Time         |


* ***Understanding of which data we are dealing with is very important while doing DATA VISUALIZATION which helps in deciding to which graph to plot.***

## 1. Univariate Analysis
**Definition:** Analysis of a single variable.

**Goal:** To understand the distribution, central tendency (mean, median), and spread (variance, standard deviation).

## 2. Bivariate Analysis
**Definition:** Analysis of two variables to find relationships or comparisons.

**Goal:** Understand correlation, trend, or impact between two variables.

## 3. Multivariate Analysis
**Definition:** Analysis of more than two variables at a time.

**Goal:** To understand the effect of multiple variables together.

| Type         | Number of Variables | Purpose                           | Example                    |
| ------------ | ------------------- | --------------------------------- | -------------------------- |
| Univariate   | 1                   | Understand distribution           | Histogram of Age           |
| Bivariate    | 2                   | Find relationships between 2 vars | Age vs Salary scatter plot |
| Multivariate | 3 or more           | Study complex relationships       | Age, Experience → Salary   |

### linestyle in Matplotlib
The linestyle argument in matplotlib.pyplot.plot() controls the style of the line in a 2D line plot.
| Style    | Code             | Description          |
| -------- | ---------------- | -------------------- |
| Solid    | `'-'`            | Default solid line   |
| Dashed   | `'--'`           | Dashed line          |
| Dotted   | `':'`            | Dotted line          |
| Dash-dot | `'-.'`           | Dash followed by dot |
| No line  | `''` or `'None'` | Hides the line       |


```python
plt.plot(batsman["index"],batsman["V Kohli"],color="#8ba901",linestyle="dashed")
plt.plot(batsman["index"],batsman["RG Sharma"],color="#008792",linestyle="-.")
plt.title("Rohit Sharma vs Virat Kohli Carrier Comparsion")
plt.xlabel("IPL Season")
plt.ylabel("Runs")
```

#### linewidth in Matplotlib
The linewidth (or lw) parameter in matplotlib.pyplot.plot() sets the thickness of the line in a 2D line plot.

```python
plt.plot(batsman["index"],batsman["V Kohli"],color="#8ba901",linestyle="dashed",linewidth=6)
plt.plot(batsman["index"],batsman["RG Sharma"],color="#008792",linestyle="-.")
plt.title("Rohit Sharma vs Virat Kohli Carrier Comparsion")
plt.xlabel("IPL Season")
plt.ylabel("Runs")
```

#### marker in Matplotlib
The marker parameter in matplotlib.pyplot.plot() is used to show symbols at each data point on a line plot. This helps you clearly see the actual data values on the line.
| Symbol | Code  | Description   |

| `o`    | `'o'` | Circle marker |
| `^`    | `'^'` | Triangle up   |
| `v`    | `'v'` | Triangle down |
| `s`    | `'s'` | Square marker |
| `*`    | `'*'` | Star marker   |

| `x`    | `'x'` | X marker      |
| `D`    | `'D'` | Diamond       |

#### Key points

## Scatter Plot

A scatter plot shows the relationship between two variables using dots. It’s ideal for visualizing correlation, trends, or clustering.
* Bivariate Analysis
* numerical vs numerical
* Use case - Finding correlation

```python
#plt.scatter simple function
x = np.linspace(-10,10,50)
#This function returns evenly spaced numbers over a specified range.np.linspace(start, stop, num=50)
y = 10*x+3+np.random.randint(0,300,50)

plt.scatter(x,y)
```

```python
#plt.scatter on pandas data
batter = pd.read_csv("01/batter.csv")
#getting top 50 batsman
batter = batter.head(50)
batter
```

```python
plt.scatter(batter["avg"],batter["strike_rate"])
for i in range(len(batter)):
    plt.text(batter["avg"][i]+0.1, batter["strike_rate"][i]+0.1, batter["batter"][i], fontsize=6)
plt.title("Strike Rate and Average Analysis of Top 50")
plt.xlabel("Average")
plt.ylabel("Strike Rate")
plt.grid()
plt.show()
```

**By doing, now we can easyly analyze the scatter plot like**
* ***at the top right we have most destructive and reliable batter.***
* ***at the bottom right we have reliable batter but not that destructive.***
* ***at the top left we have destructive batter but not that reliable.***
* ***at bottom left we have neither destructive nor reliable batter.***

#### Explanation for the function plt.text we used
* **plt.text(x, y, label) places a label at (x, y)**
* **x[i]+0.1, y[i]+0.1 slightly offsets the text so it doesn't overlap the point**

```python
plt.scatter(batter["avg"],batter["strike_rate"],color="coral",marker="*")
#s for marker size
for i in range(len(batter)):
    plt.text(batter["avg"][i]+0.1, batter["strike_rate"][i]+0.1, batter["batter"][i], fontsize=6)
plt.title("Strike Rate and Average Analysis of Top 50")
plt.xlabel("Average")
plt.ylabel("Strike Rate")
plt.grid()
plt.show()
```

### Stacked Bar Chart

A stacked bar chart shows multiple data series stacked on top of each other in a single bar, making it easy to see both the total and the part-wise contribution.

**now, we are going to make a bar charts of total runs of these players and also showing the parts to how contribution of season runs.**

## Pie Chart

* Univariate/Bivariate Analysis
* Categorical vs numerical
* Use case - To find contibution on a standard scale

```python
data1 = [23,100,49,10]
subjects = ["Eng","Phy","CP","DSA"]
plt.pie(data1,labels=subjects)
```

**How this pie chart is made?**
* **It took the sum of the array and by using the percentage of each number to the sum, it plotted the pie chart.**

```python
# dataset
gayle = pd.read_csv("01/gayle-175.csv")
gayle
```

```python
plt.pie(gayle["batsman_runs"],labels=gayle["batsman"],autopct="%.1f%%")
plt.legend()
plt.show()
```

| Format     | Output Example | Meaning           |
| ---------- | -------------- | ----------------- |
| `'%.1f%%'` | `12.3%`        | 1 decimal place   |
| `'%.0f%%'` | `12%`          | Whole number only |
| `'%.2f%%'` | `12.34%`       | 2 decimal places  |

#### colors

```python
plt.pie(gayle["batsman_runs"],labels=gayle["batsman"],autopct="%.1f%%",colors=["#00a952","#6a0000","#df8400","#2aaaf9","#00249e","#ce0a0a"])
plt.legend()
plt.show()
```

#### explode shadow

```python
plt.pie(gayle["batsman_runs"],labels=gayle["batsman"],autopct="%.1f%%",explode=[0.1,0,0,0,0,0],shadow=True)
plt.legend()
plt.show()
```

```python
plt.pie(gayle["batsman_runs"],labels=gayle["batsman"],autopct="%.1f%%",explode=[0.1,0,0.,0,0.3,0],shadow=True)
plt.legend()
plt.show()
```
