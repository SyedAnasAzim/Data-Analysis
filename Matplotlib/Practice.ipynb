# Answers:
## Answer 1

```python
x = np.linspace(-4,4,100)
y = np.linspace(-4,4,100)
x,y = np.meshgrid(x,y)
z = np.cos(x**2+y**2)

fig = plt.figure()
ax = plt.subplot(projection="3d")
var = ax.plot_surface(x,y,z,cmap="coolwarm")
fig.colorbar(var)
plt.show()
```

## Answer 3:

```python
x = np.linspace(-5,5,100)
y = np.linspace(-5,5,100)
x,y = np.meshgrid(x,y)
for t in range(6):
    z = np.sin(np.sqrt(np.maximum(0, x**2 + y**2 - t)))
    fig = plt.figure()
    ax = plt.subplot(projection="3d")
    var = ax.plot_surface(x,y,z,cmap="jet")
    fig.colorbar(var)
    ax.set_title("Surface Plot of z = sin(sqrt(xÂ² + yÂ² - t))")
    ax.set_xlabel("x")
    ax.set_ylabel("y")
    ax.set_zlabel("z = sin(sqrt(xÂ² + yÂ² - t))")
    plt.pause(2)
    plt.show()
```

## Answer 4

```python
t = np.linspace(0,4*np.pi,100)
x = np.cos(t)
y = np.sin(t)
z = t

fig = plt.figure()
ax = plt.subplot(projection = "3d")
ax.plot(x,y,z,marker=".",color="red")
ax.set_title("Line plot")
ax.set_xlabel("cos(t)")
ax.set_ylabel("sin(t)")
ax.set_zlabel("t")
plt.show()
```

## Answer 5

```python
t = np.linspace(0,4*np.pi,100)
x = t * np.cos(t)
y = t * np.sin(t)
z = t

fig = plt.figure()
ax = plt.subplot(projection = "3d")
ax.plot(x,y,z,linestyle="--",color="green")
ax.set_title("3D Spiral Curve")
ax.set_xlabel("t x cos(t)")
ax.set_ylabel("t x sin(t)")
ax.set_zlabel("t")
plt.show()
```

## Answer 7

```python
# For consistent random values
np.random.seed(42)

# Generate 3 clusters (30 points each)
x1 = np.random.normal(2, 0.5, 30)
y1 = np.random.normal(2, 0.5, 30)
z1 = np.random.normal(2, 0.5, 30)

x2 = np.random.normal(7, 0.5, 30)
y2 = np.random.normal(7, 0.5, 30)
z2 = np.random.normal(7, 0.5, 30)

x3 = np.random.normal(5, 0.5, 30)
y3 = np.random.normal(1, 0.5, 30)
z3 = np.random.normal(9, 0.5, 30)

# Plotting
fig = plt.figure()
ax = fig.add_subplot(projection="3d")

# Scatter each cluster with different color and label
ax.scatter(x1, y1, z1, c='blue', label='Cluster A', alpha=0.7)
ax.scatter(x2, y2, z2, c='red', label='Cluster B', alpha=0.7)
ax.scatter(x3, y3, z3, c='green', label='Cluster C', alpha=0.7)

# Labels and legend
ax.set_title("3D Categorized Scatter Clusters")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.set_zlabel("Z-axis")
ax.legend()

plt.show()
```

## Answer 8

```python
x = np.linspace(-5,5,100)
y = np.linspace(-5,5,100)
x,y = np.meshgrid(x,y)
z = np.sin(x) * np.cos(y)

fig = plt.figure()
ax = fig.add_subplot(projection="3d")
var = ax.plot_surface(x,y,z,cmap="Blues")
mask = z > 0.5
xx = x[mask]
yy = y[mask]
zz = z[mask]

ax.scatter(xx,yy,zz,c="red",s=1)
ax.set_title("z = sin(x) * cos(y)")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("sin(x) * cos(y)")
fig.colorbar(var)
plt.show()
```
