#### alpha Parameter in Matplotlib
The alpha parameter controls the transparency (opacity) of plot elements.

**Value Range:**
* alpha = 0 â†’ completely transparent
* alpha = 1 â†’ fully opaque
* Any float between 0 and 1 adjusts the transparency accordingly.

**Why use alpha?**
* Useful for overlaying plots to see overlaps.
* Helps reduce visual clutter(*Too much information shown at once, making the plot hard to read or understand*) in dense plots.

```python
plt.figure(figsize=(18,10))
plt.scatter(sample_df["avg"],sample_df["strike_rate"],s=sample_df["runs"])
for i in sample_df.index:
    plt.annotate(
        sample_df["batter"].loc[i],
        xy=(sample_df["avg"].loc[i],sample_df["strike_rate"].loc[i]),
        xytext=(sample_df["avg"].loc[i]+0.1,sample_df["strike_rate"].loc[i]+0.1),
        fontsize=12,color="coral",
        weight="bold"
)
plt.show()
```

## 3D Surface Plots

A 3D surface plot is used to visualize three-dimensional data. It shows a surface that connects a set of data points in 3D space using color and elevation to represent a fourth dimension (z-values based on x and y).

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)

xx,yy = np.meshgrid(x,y)
z = xx**2 + yy**2

plt.figure()
ax = plt.subplot(projection="3d")
ax.plot_surface(xx,yy,z)
```

#### np.meshgrid
numpy.meshgrid is a function that takes two 1D arrays representing x and y coordinates and returns coordinate matrices from them. These coordinate matrices are used for evaluating functions over a gridâ€”like in 3D plotting (e.g., surface plots).

```python
x = np.array([1, 2, 3])  # X-coordinates
y = np.array([10, 20])   # Y-coordinates

X, Y = np.meshgrid(x, y)

print("X:\n", X)
print("Y:\n", Y)
```

**What does it do?**
* It expands your 1D x and y arrays into 2D grids.
* Now you can calculate values (like z = f(x, y)) at each combination of x and y.

(x=1, y=10), (x=2, y=10), (x=3, y=10)
(x=1, y=20), (x=2, y=20), (x=3, y=20)

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)

xx,yy = np.meshgrid(x,y)
z = xx**2 + yy**2

plt.figure()
ax = plt.subplot(projection="3d")
ax.plot_surface(xx,yy,z,cmap="viridis")
```

## Contour Plot
### What is a Contour Plot?
**A contour plot is a 2D plot used to represent a 3D surface. It shows lines of constant value (like height, temperature, pressure, etc.) on a flat surface.**

**Think of it like a topographic map â€” the kind that shows mountains and valleys using curved lines.**
### Real-Life Analogy
**Imagine a mountain:**
* From the side, you see the height (3D surface).
* From the top (birdâ€™s eye view), you only see lines connecting points at the same height.
Those lines are contour lines.
### Why Use Contour Plots?
Contour plots are useful when:
* You want to see the shape of a surface without plotting it in 3D.
* You're working with heatmaps, elevation maps, or anything with 3 variables:
    - ***x, y (horizontal position)***
    - ***z (value/height at each point)***
### How to Read a Contour Plot
* Each line represents a constant z-value (e.g. temperature = 50Â°C).
* If lines are close together, the surface is steep.
* If lines are far apart, the surface is flat.
* The shape of lines tells you how the surface behaves.

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)

xx,yy = np.meshgrid(x,y)
z = xx**2 + yy**2

fig = plt.figure()
ax = plt.subplot(projection="3d")
p = ax.plot_surface(xx,yy,z,cmap="viridis")
fig.colorbar(p)
```

**Here, this 3D graph is converted into 2D and we are watching it from top.**

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)

xx,yy = np.meshgrid(x,y)
z = xx**2 + yy**2

fig = plt.figure()
ax = plt.subplot()
p = ax.contour(xx,yy,z,cmap="viridis")
fig.colorbar(p)
```

* Using levels parameter to increase the lines.
* The levels parameter controls how many contour lines or filled color bands you see in your plot.

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)

xx,yy = np.meshgrid(x,y)
z = xx**2 + yy**2

fig = plt.figure()
ax = plt.subplot()
p = ax.contour(xx,yy,z,cmap="viridis",levels=20)
fig.colorbar(p)
```

**fig.colorbar(p) at the will help us understand depth that which is lower and above.**

### What is a Filled Contour Plot?
* A filled contour plot is like a contour plot â€” but instead of just drawing lines, it fills the spaces between those lines with color.
* Itâ€™s a 2D way to visualize a 3D surface by using color intensity to show how high or low the values are.

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)

xx,yy = np.meshgrid(x,y)
z = xx**2 + yy**2

fig = plt.figure()
ax = plt.subplot()
p = ax.contourf(xx,yy,z,cmap="viridis")
fig.colorbar(p)
```

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)

xx,yy = np.meshgrid(x,y)
z = xx**2 + yy**2

fig = plt.figure()
ax = plt.subplot()
p = ax.contourf(xx,yy,z,cmap="viridis",levels=20)
fig.colorbar(p)
```

## Pandas plot()

### 1. Matplotlib
* Low-level plotting library (more control).
* You have to build everything manually: figure, axes, labels, etc.
* Very flexible and customizable.
* Used as the base for other libraries (Seaborn, Pandas, etc.)

### 2. Pandas .plot()
* High-level plotting (easy and quick).
* Built on top of Matplotlib.
* Automatically understands DataFrames and Series.
* Useful when working with tabular data.

| Feature          | **Matplotlib**        | **Pandas `.plot()`**        |
| ---------------- | --------------------- | --------------------------- |
| Level            | Low-level             | High-level                  |
| Flexibility      | Very high             | Medium                      |
| Ease of use      | More complex          | Very simple                 |
| Works best with  | Any kind of data      | DataFrames / Series         |
| Customization    | Full control          | Limited, but can use kwargs |
| Speed of writing | Slower                | Faster                      |
| Base library for | Pandas, Seaborn, etc. | Depends on Matplotlib       |
